```python
# Import pandas 
import pandas as pd
```

```python
# read the data
data = pd.read_csv('https://raw.githubusercontent.com/analyticsindiamagazine/MocksDatasets/main/Iris.csv')
```

```python
# Top 5 rows
data.head()
```

```python
# check the shape of data
data.shape
```

```python
# Defining input (X) features
X = data.iloc[:,1:5].values
```

```python
# size of X
X.shape
```

#**Initializing and training the Gaussien mixture models**

```python
# check the distribution
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1])
plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length')
```

```python
# performing GMM
from sklearn.mixture import GaussianMixture
GMM = GaussianMixture(n_components=3)
# train model
pred = GMM.fit_predict(X) 
```

#**Predicting and Plotting the clusters**

```python
# Predicting the probability
GMM.predict_proba(X[:11,:])
```

```python
# predicted result
data['labels'] = pred
```

```python
# Cluster label
data['labels'].value_counts()
```

```python
import matplotlib.pyplot as plt
# plotting the clusters 
plt.scatter(X[pred == 0, 0], X[pred == 0, 1], s = 50, c = 'green', label = 'Setosa')
plt.scatter(X[pred == 1, 0], X[pred == 1, 1], s = 50, c = 'yellow', label = 'Versicolor')
plt.scatter(X[pred == 2, 0], X[pred == 2, 1], s = 50, c = 'black', label = 'Virginica')
plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length')
plt.title('Identified iris clusters')
plt.legend()
```
