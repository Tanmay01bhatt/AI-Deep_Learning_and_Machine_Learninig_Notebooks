```python
# Importing 
from gluoncv import model_zoo, data, utils
from matplotlib import pyplot as plt
import tensorflow as tf
```

# **Initaiting the YOLO**

```python
# load the moel
net = model_zoo.get_model('yolo3_darknet53_voc', pretrained=True)
```

```python
from IPython.core.pylabtools import figsize
def detect_object(url,name):
  get_image = tf.keras.utils.get_file(name + '.jpeg',url)
  x, img = data.transforms.presets.yolo.load_test(get_image, short=512)
  print('Shape of pre-processed image:', x.shape)
  class_IDs, scores, bounding_boxs = net(x)
  plt.figure(figsize=(14,7))
  ax = utils.viz.plot_bbox(img, bounding_boxs[0], scores[0],
                         class_IDs[0], class_names=net.classes)
  plt.axis('off')
  plt.show()
```

```python
#Detecting objects in the first image
detect_object('https://analyticsindiamag.com/wp-content/uploads/2022/06/object1.jpeg','Object')
```

```python
# Detecting objects in the second image
detect_object('https://analyticsindiamag.com/wp-content/uploads/2022/06/object-2.jpeg','object2')
```

```python
#Detecting objects in the third image
detect_object('https://www.mymove.com/wp-content/uploads/2020/08/Abby-Dauchess.jpg','object3')
```
