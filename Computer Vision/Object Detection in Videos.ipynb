# **Loading the YOLO model**

As we are going to use the YOLO model to detect objects, we will download its pre-trained weights.

```python
# Load weights
import tensorflow as tf
# Download the YOLO model
get_model = tf.keras.utils.get_file('yolo.h5','https://github.com/OlafenwaMoses/ImageAI/releases/download/1.0/yolo.h5')
```

# **Initiating the Video Detection**

Now, we will import the ImageAI library that has been installed. We will also import the VideoObjectDetection class

```python
from imageai.Detection import VideoObjectDetection
```

```python
vid_obj_detect = VideoObjectDetection()
vid_obj_detect.setModelTypeAsYOLOv3()
vid_obj_detect.setModelPath('/root/.keras/datasets/yolo.h5')
vid_obj_detect.loadModel()
```

# **Detecting objects in video**

```python
video_file = tf.keras.utils.get_file('traffic_vid2.mp4','https://analyticsindiamag.com/wp-content/uploads/2022/06/traffic_vid2.mp4')
detected_vid_obj = vid_obj_detect.detectObjectsFromVideo(
   input_file_path =  video_file,
   output_file_path = '/root/.keras/datasets/video_detect_res.mp4',
   frames_per_second=15,
   log_progress=True,
   return_detected_frame = True)
```
