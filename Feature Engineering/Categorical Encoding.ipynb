#**Label Encoding**

```python
# Counting category labels
data['Income Group'].value_counts()
```

```python
# Library for ordinal / label encoding
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Ordinal / label encoding
ord_enc = OrdinalEncoder()
data["Income Group"] = ord_enc.fit_transform(data[["Income Group"]])
data
```
