So far know we have discussed that Named entity recognition is a process where the named entity gets identified and linked to its class. As we know that any given raw text data consists of various kinds of words some of them are stopwords, and part of speech words likewise there can be various kinds of words that can be presented in a text file which can be segregated as named entities.

So the NER can be implemented using many ways out of those two popularly used libraries are NLTk and Spacy. Spacy is an open-source NLP library that provides various facilities and packages which can help full-on NLP tasks such as POS tagging, lemmatization, and fast sentence segmentation. The following steps are to be taken to implement NER using the SpaCy library following steps can be taken.  



```python
import spacy
from spacy import displacy
# load the enviroment for english
nlp = spacy.load('en_core_web_sm')#small eng pipeline trained on web texts
```

# **Set the data**

```python
# data from which entities to be recognised 
raw_text="""Bengaluru (also called Bangalore) is the capital of India's southern Karnataka state. 
The center of India's high-tech industry, the city is also known for its parks and nightlife. 
By Cubbon Park, Vidhana Soudha is a Neo-Dravidian legislative building. 
Former royal residences include 19th-century Bangalore Palace, modeled after England’s Windsor Castle, and Tipu Sultan’s Summer Palace, an 18th-century teak structure."""
```

# **Grabbing the entities**

```python
# detecting the entities
text = nlp(raw_text)
```

Fitting the model on the sample text.

```python
# printing the individual entity 
for w in text.ents:
    print(w.text,w.label_)
```

# **Displaying and explaining the entities**

We can also visualize the name entities with the data using the displacy package of spacy.

```python
# Highlighting the entitties 
spacy.displacy.render(text, style="ent",jupyter=True)
```

Here one thing we can get confused about is the named entity code. We can also check for the explanation of those NE codes using the method spacy.explain() for each class of the entity.

```python
spacy.explain('ORG')
```

```python
spacy.explain('GPE')
```

```python
spacy.explain('FAC')
```
