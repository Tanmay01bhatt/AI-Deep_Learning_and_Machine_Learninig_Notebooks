# **Visual analysis of data**

Now we will start the visual analysis by first plotting the histogram for all the variables (continuous and numeric categorical variables).

```python
# Histogram of all columns
data.hist(figsize=(20,10))
```

```python
 # Histogram of one columns
data['Months_on_book'].hist()
```

```python
data.hist(column='Months_on_book', bins=10, grid=False, color='magenta', zorder=2, rwidth=0.5)
```

You can see the histogram of the above two columns and the distribution of data looks like Gaussian distribution.

Next histogram for categorical data can be obtained as follows

```python
# Histogram of categorical data
data['Dependent_count'].value_counts().plot(kind='bar', color = 'red', figsize=(8,6), xlabel = 'Dependents', ylabel = 'Frequency')
```

```python
# Histogram of categorical data
data['Education_Level'].value_counts().plot(kind='bar', color = 'red', figsize=(8,6), xlabel = 'Dependents', ylabel = 'Frequency')
```

For categorical data apart from the histogram, the Pie chart can also bring more information, by looking at the pie chart we can also conclude about the contribution of each category.

```python
#pie chart of categorical data
data['Card_Category'].value_counts().plot(kind='pie', figsize=(7, 7), legend=True, title='Card Categories')
```

When it comes to analysing the outliers one should use the box plot, by looking at the box plot we can conclude about outliers.

```python
# Simple box plot (all columns)
data.plot.box()
```

Above we have plotted a box plots for all variables, but we can infer any from the above chart. Lets obtain a boxplot for the single column as below:

```python
# Simple box plot (for one column)
data['Months_on_book'].plot.box()
```

As we can see the data is nearly normally distributed and have some outliers at both tails.

Now finally to analyse the trend between the contentious variables the scatter plot is used which can be obtained as follows:

```python
# scatter plot between experience and salary
data.plot.scatter(x='Customer_Age', y='Months_on_book', figsize=(12, 7), color='blue', title='Age Vs Months_on_Book')
```

Another method used to analyse the effect of one variable on the rest of the other variables is obtaining a correlation table. Below we are obtaining a correlation table for all columns.

```python
# Correlation analysis
data.corr(method='pearson', min_periods=1)
```

Above we can see in the table the effect of customer_age on months_on_book is 0.78 means it is positively correlated by a factor of 0.78.

Finall,y we can obtain the covariance matrix as below,

```python
# Covariance
data.cov(min_periods=None, ddof=1)
```
