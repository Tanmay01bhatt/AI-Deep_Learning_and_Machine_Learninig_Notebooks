```python
# pandas for Data handling
import pandas as pd

# Numpy for numerical operations 
import numpy as np
```

```python
# movies title dataset
movies_title = pd.read_csv('https://gitlab.com/AnalyticsIndiaMagazine/practicedatasets/-/raw/main/bootcamp/recommender_system/movies.csv')
movies_title.head()
```

```python
# movies ratings dataset
movies_ratings = pd.read_csv('https://gitlab.com/AnalyticsIndiaMagazine/practicedatasets/-/raw/main/bootcamp/recommender_system/ratings.csv')
movies_ratings.head()
```

```python
# Merging the titles and ratings
title_ratings = pd.merge(movies_title,movies_ratings)
```

```python
# Visualize top 5 rows
title_ratings.head()
```

```python
# Check the shapes
title_ratings.shape
```

#**Prepare the data for recommendations**

```python
# Droping the irrelevant columns
title_ratings.drop(['genres','timestamp'], axis=1,inplace=True)
```

```python
# Pivote table
UserRatings = title_ratings.pivot_table(index=['userId'],columns=['title'],values='rating')
UserRatings.head()
print("Before: ",UserRatings.shape)
UserRatings = UserRatings.dropna(thresh=10, axis=1).fillna(0,axis=1)
print("After: ",UserRatings.shape)
UserRatings.head()
```

#**Building correlation metrix**

```python
# Pearson correlations
relation_metrix = UserRatings.corr(method='pearson')
relation_metrix.head()
```

```python
# User defined function to reccoment the movie
def get_similar(movie_name,rating):
    similar_ratings = relation_metrix[movie_name]*(rating-2.5)
    similar_ratings = similar_ratings.sort_values(ascending=False)
    return similar_ratings
```

#**Generating the similar movies**

```python
# getting similar movie
movies= [("Skyfall (2012)",5),("Mission: Impossible III (2006)",4)]
similar_movies = pd.DataFrame()
for movie,rating in movies:
    similar_movies = similar_movies.append(get_similar(movie,rating),ignore_index = True)
```

```python
# Top 10 movies that are similar to queries
similar_movies.sum().sort_values(ascending=False).head(10)
```
